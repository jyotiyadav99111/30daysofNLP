# Recurrent Neural Network (Part -2)

This segment describes the backpropagation of the entire RNN model. Backpropagation is when the final loss is distributed across the entire weights and updated with a function of loss. The more the number of iterations more it gets closer to the actual values. But this also leads to overfitting if the iterations are too much. Let’s see how the backpropagation works in the case of RNN. Before that, we will be going through a few essential topics which are required to be understood. 

### Stochastic Gradient Descent 

Gradient descent is one of the methods to update the weights of a neural network. The visual depiction of the process is as follows:



Loss is a function of the gap between the actual value and the predicted value. It is natural to look for a weight that minimizes the loss as it will lead to the difference between the actual and predicted values. 

**Loss = f(y_hat - y)**

The minimum value of the loss function has been depicted in the graph presented above. Now, the weights are initialized which simply means that the initial values for weight have been selected randomly (one of the methods). 

As you can see, to reach a minimum the weight has to move in the opposite direction of that of a slope. This implies that some fraction of the slope will be subtracted from the old weight to adjust it towards the minimum. Therefore, the formula becomes:

**w_new = w_old - (eta)* (dL/dw_old)**

Where,

**L** is the loss function, 

**w_old** is the old weight

**w_new** is the updated weight

**eta** is the learning rate

The higher the learning rate, the more frequently the weight changes, and there are chances of it never reaching the minimum point. Whereas, a lower learning rate will require more epochs which in turn will lead to overfitting. Therefore, their learning rate is supposed to be chosen cautiously. 

### Backpropagation

Backpropagation is based on the gradient descent algorithm. The intent is the same which is to reach the minimum of the loss function. To update the weights, we require the derivative of the loss function. As depicted in the previous article, the loss function depends on the output of the previous layer as well. Therefore, the formula for the derivative will change a little. 

therefore, the chain rule will be applied here:

**Loss = f(y_hat - y)

     = f(h(x, w, o) - y)

**dL/dw = (dL/do) * (do/dw)

The above formula will be used for weight updates through gradient descent. There are many ways to initialize the weights other than random initialization.
