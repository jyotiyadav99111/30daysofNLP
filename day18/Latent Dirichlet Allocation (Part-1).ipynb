# Latent Dirichlet Allocation (Part-1)

We have already been talking about the text-preprocessing techniques, let’s dive a little deeper into this domain. The next big thing is a very frequently used word: **LDA**. Due to the complexity of the term, people end up finding it very difficult to understand. I will try to make this process as simple as possible from my end. 

The process is used mostly for supervised learning models but it doesn’t lose its power when it comes to unsupervised problems. It can be used for topic prediction as well as document clustering. 

### The analogy to human choices

Let’s consider an example of book stores. Whenever we look for books to purchase, we tend to read a few pages of them to see if that suits our taste or not. Now the question is how are we able to decipher the genre attached to any book. The answer comes from the fact that we try to identify the topics in the book through the words used there. For example, a marvel comic will have topics related to mystery, superheroes, extraordinary powers, etc. WHereas a fashion magazine will have topics related to clothing, makeup, accessories given the fact that we ignore photos in both the case. 

### Logic Behind the algorithm

In simple terms, the entire process can be divided into two subparts. One just has to remember these two points to make sense out of the logic behind them. These are mentioned below:

1. A document can be composed of a few topics simultaneously
2. And one topic can have multiple words associated with it

LDA is based on another familiar word that most of us must have heard about: **Bag-of-words**. The name is self-explanatory, it simply looks at the frequency of words, not on the position of the word in the text. As the task is of topic modeling, this assumption is fair to make to some extent. Generally, the type of words used in a text help decipher the context it is talking about. 

### Parameters of the model

There are primarily three parameters that can be considered for the algorithm at hand. These are namely: alpha, Beta, and k.

1. **Alpha:** The parameter depicts the similarity between the documents. A higher value depicts that the documents share many topics. 

2. **Beta:** The parameter depicts that the similarity in the topics themselves. A higher value implies topics share a lot of words. 

3. **K:** It depicts the number of topics that can be chosen for the modeling purpose.
