# BERT (Part-2)

The paper released by Google shows two architectures of BERT:

1. Base: It is consisting of 12 encoder layers, 12 attention heads, and 110M parameters.

2. Large: It consists of 24 encoder layers, 16 attention heads, and 340M parameters.

The abstract of the paper is as follows:

"We introduce a new language representation model called BERT, which stands for Bidirectional Encoder Representations from Transformers. Unlike recent language representation models, BERT is designed to pre-train deep bidirectional representations from unlabeled text by jointly conditioning on both left and right context in all layers."

### Advantages of using BERT

1. The model is pre-trained on a huge dataset of unlabelled text. This gives an added advantage of the lesser requirement of the data to train. The model is required to be fine-tuned to the new dataset.


2. There will be a situation where the same word is used in different contexts. For example, 'bank' can be used as a bank of the river or as an institution. BERT can resolve this using bi-directional training in the model.

Note: BERT model is a stack of encoders only.
