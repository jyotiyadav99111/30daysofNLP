# Introduction to NLP

NLP has become a very popular subject in the field of Data Science in recent years. The entire domain of NLP is a more deeper and complex than it seems from the outside. This not only involves the fancy algorithms but also some hidden hard worked string manipulations. The essential part of NLP is to make the machine understand linguistics.

### NLP Types

NLP can be implemented in broadly three ways:
1.	**Rule based Solutions:** This method is employed when the input to the algorithm is fixed or of the similar kind. For example, a chat assistant which provides you with specific questions you can choose from and gives you the similar output. Another example is information extraction from the text like some IDs, emails, and contact numbers. The common link here is the invariability in the input to the algorithm. 
2.	**Machine Learning Solutions:** Easiest example would be a spam detector. The process under this domain involves feature extraction from the text in form of numbers which serve as the input in the classic machine learning algorithms like Naïve Bayes. 
3.	**Deep Learning Solutions:** This arena has observed a lot of advancements recently. With the introduction of transformers, the entire base for NLP model construction has changed. The first example to this advancement would be a Language Translator. 


### Examples of NLP Use Cases:

1.	**Name Entity Recognition:** The task involves the extraction of person, places, organizations, etc from the text. These are later linked to form the entire network.
2.	**Sentiment Analysis:** It is one of very basic use of NLP which is to identify the polarity of the text. For example, analyzing the reviews in order to see if the customer is happy about the product they are selling or they didn’t like it. 
3.	**Language Translation:** We all have seen the language translator that Google provides. There the machine is able to convert any queries from one language to another. This is made possible through NLP. 
4.	**Chatbots:** Everyone is quite fascinated by the advancements in this area. Siri, Alexa, Cortona are a few very famous tools based out of it. They can answer any of your queries and question can be as random as possible but in a specific language. 
5.	**Summarization:** The task of summarizing a big document into a few lines. Here are two types of text summarization: Extractive and Abstractive. The former is applied when a key take always are to be extracted from the entire piece of text (important line from the text). The later is used in order to get jest of the text (New words and sentences are formulated).
6.	**Sentence Completion:** This is an interesting application. In case you want to write a story or are stuck at some place, this tool might be of use. Recently, software has been built over the existing literature that it is able to write its own stories.
