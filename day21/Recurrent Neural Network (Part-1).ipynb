# Recurrent Neural Network (Part-1)

In the entire series of NLP, we have come across many techniques like TF-IDF, word2vec, BoW. These techniques are useful in converting the entire text into some vectors. They have some advantages but disadvantages as well. The biggest disadvantage of this is their ignorance of the sequence of words in the text presented. As we are losing out on the information (sequence of the words), this creates an issue for the accuracy and might end up decreasing that for example the topic modeling or sentiment analysis. Let’s see how this can information can be captured through RNN. 

### Example 

Suppose a person goes out every night. The first day he went to cafe1, the second day to cafe2 and the third day to cafe3. The other information here is that these are the only places he goes to. Now, the data goes like below:

Cafe1, cafe2, cafe3, cafe1, cafe2, cafe3, cafe1,...

To predict where will he go today, I only need to see the cafe he visited yesterday. Therefore the sequence of the data mattress here. Unlike BoW which will only collect this information and convert it into a vector with frequency.

To further dive into the architecture of RNN, let’s take an example of sentiment analysis. The task at hand is to tag if a text/review on any app is positive(1) or negative(0). Therefore, the input sequence(review) will look like this: 

I loved this product

It can be written in terms of the input sequence as:

x11 , x12, x13, x14

Given this, the unfolded structure of the RNN will look like below:


    
As it is evident that every time stamp one input (xit) out of the review goes in and produces an output ot and then the output is transferred to the next hidden layer set. Therefore, as in the normal neural networks, the output out of a layer is a function of input and weights only, there it will be a function of the output from the previous layer as well to maintain the sequence in the text. Therefore,

o1 = f(x11 * w)

o2 = f(x12 * w + o1 * w1)

o3 = f(x13 * w + o2 * w1)

o4 = f(x14 * w + o3 * w1)

Once the entire sequence has been entered, the final output is taken and the “softmax” function has been applied to it. As the output of the entire analysis should be 0 or 1, it requires the application of the sigmoid function to do so. 

The weights are not changing in this layer because weights will be updated in the backpropagation only and derivatives are required to be derived for the same.
