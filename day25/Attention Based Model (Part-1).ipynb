# Attention Based Model (Part-1)

In the previous articles, we have gone through some of the text mining and preliminary methods for text analysis and modeling. But, these techniques are not useful when it comes to heavy models as they require a lot of text for analysis. Therefore, a new model is required for the heavy processing part of it.

In the past few years, the NLP domain has experienced a lot of surge in innovation. More and more complex models have been released by Google and other companies. One of these topics is "Attention Mechanism". Before, the release of this model in 2015, NLP was dominated by RNN/LSTM decoders and encoders.

### Problem with the Long Sequences

Taking the example of a machine translation, the following architecture is a very simple illustration of the entire procedure:



As depicted above, the task at hand is to convert a sentence in French into the English language. In layman terms, the encoder reduces the size of the sentences and converts them into a value, and stores that. And the decoder takes that value as the input and converts that into the English language.

Now, the problem that arises is when the length of the sentence is very long it leads to poor performance as it will have to remember the entire sentence at once. After almost 30-35 words sentence this complexity arises and thereby creates an issue.

### Solution: Attention is all you need

It is a model composed of encoders and decoders. The layer-wise architecture is presented below of the model:



Now, how does the input goes into the encoder layer. In the encoder layer, instead of memorizing the entire sentence at once, each word is entered one by one and an output is produced by it. The output from the previous layer is then added to the next layer with the one word and so on.

The decoder layer converts the output from all the encoder layers to translate every word into the English language. This process is repeated for all the words.
