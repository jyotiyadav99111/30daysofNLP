# TF-IDF

The text falls under the categorization of unstructured data. This implies a lack of a predefined number of features or characters in two texts. One sentence could be longer than another and there is no way to control the number of words in a sentence. Therefore, it becomes very difficult for traditional machine learning algorithms to make sense out of it. To feed the traditional algorithms, we require to extract features out of the text at hand.

One popular method is TF-IDF. It stands for **"Term Frequency-Inverse Document Frequency"**. In simple words, it assigns weights to each word occurring in a sentence. The weight is thereby calculated based on how specific it is to the document.

### Applications

It is an information retrieval method and can be used in any model for the same purpose. Mostly it is used for document classification and search optimization.

1) **Document classification**: Let's say you have two categories in which you would want to classify the documents: Sports & Science. The articles will have a lot of similar words into which are general, but what makes them different is the rarely occurring words. For example, Sports doc has words like game names, players, etc. Whereas, science will have words like atoms, research, etc. Therefore, the most frequent words are identifying a sentence, but the less occurring words end up making it different from others.

2) **Search Optimisation**: The search engine is better if it can fetch the most relevant results from the pool of documents available online. Therefore, the words one entered in the search box should be showing the highest importance (as calculated by TF-IDF) in the document which comes at the top.

### Theory and formulas

The term consists of the following two components:

1. **Term frequency**: Number of times a given word is occurring in the document. But this frequency therm is highly misleading as very common words might occur a lot of times. This is not it, it is also dependent on the length of the document. The longer the document the more times a common word will occur in that. Therefore, the frequency is normalized by dividing it by the total count of the terms in a given document.

2. **Inverse Document Frequency**: As described earlier, the uniqueness of the document is also dependent on the fact of how unique are the words that make it. In simple terms the informativeness of the word. Therefore, the lesser the number of documents containing the word, the more informative it is in that domain. Document frequency implies the fraction of documents containing a particular word. It doesn't take into consideration how many times it occurs in each document. And, the inverse of document frequency is IDF.
